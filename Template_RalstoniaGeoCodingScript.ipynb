````python
subset_df = df.iloc[:11000]
subset_df
````

````python
count = 0
lat = []
long = []

for index, row in subset_df.iterrows():
    if count > 11000:
        break
    params = {
        'q': row['Location Isolated'],
        'access_token': access_token
    }
    # Make the request
    response = requests.get(base_url, params=params)
    count += 1
    # Check if the request was successful
    if response.status_code == 200:
        # Parse the JSON response
        data = response.json()
        # Check if coordinates are present in the response
        if data.get('features'):
            coordinates = data.get('features')[0].get('geometry').get('coordinates')
            print(f"{row['Location Isolated']}'s coordinate are {coordinates}")
            lat.append(coordinates[0])
            long.append(coordinates[1])
        else:
            # Handle cases where coordinates are not found
            lat.append(None)  # or any suitable placeholder
            long.append(None)
            print(f"Coordinates not found for {row['Location Isolated']}")
    else:
        print(f"Error: {response.status_code}")
        # Append placeholders for failed requests to maintain length consistency
        lat.append(None)
        long.append(None)

# Now 'lat' and 'long' will have the same length as 'subset_df'
subset_df["Long"] = long
subset_df["Lat"] = lat
subset_df.head()
````

````python
count = 0
lat = []
long = []

for index, row in subset_df.iterrows():
    if count > 11000:
        break

    location = row['Location Isolated']

    # Check if the location is empty
    if pd.isna(location) or location.strip() == "":
        lat.append(None)
        long.append(None)
        print(f"Skipped empty location at index {index}")
        continue  # Skip to the next iteration

    params = {
        'q': location,
        'access_token': access_token
    }

    # Make the request
    response = requests.get(base_url, params=params)
    count += 1

    # Check if the request was successful
    if response.status_code == 200:
        # Parse the JSON response
        data = response.json()
        # Check if coordinates are present in the response
        if data.get('features'):
            coordinates = data.get('features')[0].get('geometry').get('coordinates')
            print(f"{location}'s coordinates are {coordinates}")
            lat.append(coordinates[1])  # longitude is at index 0
            long.append(coordinates[0])  # latitude is at index 1
        else:
            # Handle cases where coordinates are not found
            lat.append(None)  # or any suitable placeholder
            long.append(None)
            print(f"Coordinates not found for {location}")
    else:
        print(f"Error: {response.status_code}")
        # Append placeholders for failed requests to maintain length consistency
        lat.append(None)
        long.append(None)

# Now 'lat' and 'long' will have the same length as 'subset_df'
subset_df["Long"] = long
subset_df["Lat"] = lat
subset_df.head()
````
